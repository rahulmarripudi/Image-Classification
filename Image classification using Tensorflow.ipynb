```python
import os
import splitfolders
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import pathlib
# Image Resize
img_height, img_width = 200, 200
input_shape = (img_height, img_width, 3)
batch_size = 16

# Split
data_bs = r'C:\Users\USER\Downloads\Image_classification_k\Quad Leaders'
data_bs = pathlib.Path(data_bs)
data_bs
splitfolders.ratio(data_bs, output='Image_classification_k/op/', ratio=(0.7, 0.15, 0.15), seed=123)
```

```python
# This code defines BaseModel.The model is built using several convolutional, pooling, activation, dropout, and dense layers. 
# The call method specifies the forward pass of the model.
```

```python
#x is a variable.i create a cnn model with 7 layers.
```

```python


# imagecl/op means the original dataset with 4 sub folders is split into train,test and val .
# tra,test,val have 4 pm images


# Image Rescaling
datagen = ImageDataGenerator(rescale=1.0 / 255)

# # Training , Testing and Validating
train_ds = datagen.flow_from_directory(r'C:\Users\USER\Downloads\Image_classification_k\Image_classification_k\op\train', target_size=(img_height, img_height),
                                       class_mode='sparse', batch_size=batch_size)

test_ds = datagen.flow_from_directory(r'C:\Users\USER\Downloads\Image_classification_k\Image_classification_k\op\test', target_size=(img_height, img_height),
                                      shuffle=False, class_mode='sparse', batch_size=batch_size)

val_ds = datagen.flow_from_directory(r'C:\Users\USER\Downloads\Image_classification_k\Image_classification_k\op\val', target_size=(img_height, img_height),
                                     shuffle=False, class_mode='sparse', batch_size=batch_size)
#we read the images in train test and validation folders.the images are set to image width,height and 
#batsize parameters set by us.

#base model is cnn model.I create 7 layers in it.x is  a variable.call is a forward pass of the model.
#
# # Model
class BaseModel(tf.keras.Model):
    def __init__(self):
        super(BaseModel, self).__init__()
        self.C1 = tf.keras.layers.Conv2D(32, (3, 3), padding='same', input_shape=input_shape)
        self.B1 = tf.keras.layers.BatchNormalization()
        self.A1 = tf.keras.layers.Activation('relu')
        self.P1 = tf.keras.layers.MaxPooling2D(2, padding='same')
        
        self.C2 = tf.keras.layers.Conv2D(32, (3, 3), padding='same')
        self.B2 = tf.keras.layers.BatchNormalization()
        self.A2 = tf.keras.layers.Activation('relu')
        self.P2 = tf.keras.layers.MaxPooling2D(2, padding='same')
        self.Dr1 = tf.keras.layers.Dropout(0.2)
        
        self.F1 = tf.keras.layers.Flatten()
        self.D1 = tf.keras.layers.Dense(128, activation='relu')
        self.D2 = tf.keras.layers.Dense(128, activation='relu')
        self.D3 = tf.keras.layers.Dense(128, activation='relu')
        self.Dr2 = tf.keras.layers.Dropout(0.2)
        self.D4 = tf.keras.layers.Dense(5, activation='softmax')
        
    def call(self, x):
        x = self.C1(x)
        x = self.B1(x)
        x = self.A1(x)
        x = self.P1(x)
        
        x = self.C2(x)
        x = self.B2(x)
        x = self.A2(x)
        x = self.P2(x)
        x = self.Dr1(x)
        
        x = self.F1(x)
        x = self.D1(x)
        x = self.D2(x)
        x = self.D3(x)
        x = self.Dr2(x)
        y = self.D4(x)
        return y
    
# Compilation
model = BaseModel()
model.compile(optimizer= tf.keras.optimizers.legacy.Adam(),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['sparse_categorical_accuracy'])
#Compilation means training the model
# Setup
checkpoint_save_path = './BaseModelForImgs.ckpt'
if os.path.exists(checkpoint_save_path + '.index'):
    print('---------loading---------')
    model.load_weights(checkpoint_save_path)
#base model means dl model for images in imgs folder
#checkpoint file consists of acc and loss every epoch.
cp_callback = tf.keras.callbacks.ModelCheckpoint(filepath=checkpoint_save_path,
                                                 save_best_only=True, save_weights_only=True)

# Training of model
history = model.fit(train_ds, epochs=20, batch_size=batch_size, 
                    validation_data=val_ds, callbacks=[cp_callback])

# Model summary
model.summary()

# Saving of model
file = open('./BaseModelWeights.txt', 'w')
for v in model.trainable_weights:
    file.write(str(v.name) + '\n')
    file.write(str(v.shape) + '\n')
    file.write(str(v.numpy()) + '\n')
file.close()
# #For every iteration,it writes the eval metrics to the file.
# #dl model gets trained and we store the evaluation metrics in the file.
# #
```

```python
import matplotlib.pyplot as plt

acc = history.history['sparse_categorical_accuracy']
loss = history.history['loss']

plt.figure(figsize=(12, 6))
plt.style.use('ggplot')
plt.subplot(1, 2, 1)
plt.plot(acc, label='Training Accuracy')
plt.title('BaseModel Training Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(loss, label='Training Loss')
plt.title('BaseModel Training Loss')
plt.legend()

plt.show()
```
